```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "/home/rahul/github_projects/kh_analytics_interview/data/raw/amazon.csv"
)
```

```python
df.head()
```

```python
df["actual_price"] = (
    df["actual_price"].str.replace("‚Çπ", "").str.replace(",", "").astype(float)
)
df["discounted_price"] = (
    df["discounted_price"].str.replace("‚Çπ", "").str.replace(",", "").astype(float)
)
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "").astype(float)
df["rating"] = df["rating"].str.replace("|", "0").str.replace("''", "0").astype(float)
df["rating_count"] = df["rating_count"].str.replace(",", "").astype(float)
```
